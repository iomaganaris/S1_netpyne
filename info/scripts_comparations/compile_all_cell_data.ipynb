```python
import os
import neuron   

rootFolder = '/home/fernando/S1_BBP/'
os.chdir(rootFolder)

folder = os.listdir('cell_data/')
folder = sorted(folder)
print(folder)

compilemods = 1
copyStochKv_deterministic = 1
```

```python
## https://bbp.epfl.ch/nmc-portal/assets/documents/static/Download/hoc_combos_syn.1_0_10.allzips.tar
## extracted 1035 folders in home/fernando/S1_BBP/cell_data/  #~ not inclued in the github

copyStochKv_deterministic = True
compilemods = True

StochKvcells = []
nonStochKvcells = []

for cellName in folder:
    os.chdir('/home/fernando/S1_BBP/cell_data/'+cellName+'/')
    
    foldermech = os.listdir('mechanisms/')
    if 'StochKv.mod' in foldermech:
        StochKvcells.append(cellName) 
    else:
        nonStochKvcells.append(cellName) 
```
